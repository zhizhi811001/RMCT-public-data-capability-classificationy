This notebook documents the conceptual-to-operational construction workflow for RMCT. It is designed for transparency and reproducibility. The public repository does not redistribute full raw website texts; where full text is required, researchers should run the scripts on an authorised local corpus.

## 1. Inputs

Expected public inputs:

- `data/rmct_controlled_vocabulary.csv`
- `data/public_company_frame_sample.csv`
- `data/rmct_company_profiles_anonymised.csv`

Optional private/local inputs for full reproduction:

- `data/private/website_texts.csv` with `company_id`, `url`, `crawl_date`, `raw_text`
- literature concept extraction file, if re-running the literature clustering stage

````python
from pathlib import Path
import pandas as pd

DATA = Path('../data')
vocab_path = DATA / 'rmct_controlled_vocabulary.csv'
profiles_path = DATA / 'rmct_company_profiles_anonymised.csv'

vocab = pd.read_csv(vocab_path)
profiles = pd.read_csv(profiles_path)

print('Vocabulary rows:', len(vocab))
print('Company profile rows:', len(profiles))
display(vocab.head())
display(profiles.head())
````

## 2. Vocabulary audit

The controlled vocabulary links heterogeneous website expressions to preferred labels, narrower capability categories and broader RMCT dimensions.

````python
dimension_summary = (
    vocab.groupby('broader_dimension', as_index=False)
    .agg(
        n_preferred_labels=('preferred_label', 'nunique'),
        n_narrower_categories=('narrower_category', 'nunique'),
        total_mentions=('total_mentions', 'sum')
    )
    .sort_values('n_preferred_labels', ascending=False)
)
display(dimension_summary)
````
