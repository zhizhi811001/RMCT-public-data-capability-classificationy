# Website-derived concept clustering validation

This notebook documents the independent empirical cross-check between website-derived capability concepts and RMCT dimensions. Full website text is not redistributed; the notebook therefore focuses on reproducibility logic and expected public outputs.

## 1. Method summary

The empirical validation stage independently extracts capability concepts from public manufacturer website text, selects diverse concepts for clustering, embeds phrases using Sentence-BERT, reduces dimensions using UMAP and identifies clusters using HDBSCAN. Website-derived clusters are then cross-checked against RMCT dimensions.

```python
from pathlib import Path
import pandas as pd

DATA = Path('../data')
profiles = pd.read_csv(DATA / 'rmct_company_profiles_anonymised.csv')

dimension_cols = [c for c in profiles.columns if c.endswith('Capability') or c in [
    'Commercial & Supply Readiness',
    'Sustainability, Circularity & Lifecycle Management Capability'
]]

print('Company profiles:', len(profiles))
print('Detected dimension columns:', dimension_cols)
display(profiles[dimension_cols + ['n_dimensions_detected']].head())
```

## 2. Dimension visibility check

This table reports how many anonymised companies have evidence for each RMCT dimension in the public profile file.

```python
coverage = []
for col in dimension_cols:
    coverage.append({
        'rmct_dimension': col,
        'n_companies_with_evidence': int((profiles[col] > 0).sum()),
        'coverage_percent': round((profiles[col] > 0).mean() * 100, 2)
    })
coverage = pd.DataFrame(coverage).sort_values('coverage_percent', ascending=False)
display(coverage)
```
